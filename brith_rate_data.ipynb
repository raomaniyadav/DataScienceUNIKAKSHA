```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\unikaksha\\datascience\\project_DS\\miniProject4\\birth.csv")
```

```python
data.head()
```

```python
#Create pivot table with index- decade and column 'gender'
```

```python
data["decade"]=10*(data["year"]//10)
```

```python
data.head()
```

```python
data_p=pd.pivot_table(data,index="decade",columns="gender",aggfunc="sum")
```

```python
data_p
```

```python
sns.set()
data_p.plot()
plt.ylabel('total births per year')
plt.show()
```

```python
data_p.plot(kind="bar")
plt.ylabel('total births per year')
plt.show()
```
